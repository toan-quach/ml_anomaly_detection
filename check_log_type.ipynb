```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/preprocessed/Event_traces.csv")
```

```python
failed_data = data[data["Label"] == "Fail"].drop_duplicates(subset=['Features'])
success_data = data[data["Label"] == "Success"].drop_duplicates(subset=['Features'])
```

```python
failed_data["Features"] = failed_data["Features"].map(lambda x: x[1:-1].split(","))
success_data["Features"] = success_data["Features"].map(lambda x: x[1:-1].split(","))
```

```python
failed_operation_count_unique = {}
failed_data["Features"].map(lambda x: [failed_operation_count_unique.update({op: failed_operation_count_unique.get(op, 0) + 1}) for op in set(x)])
success_operation_count_unique = {}
success_data["Features"].map(lambda x: [success_operation_count_unique.update({op: success_operation_count_unique.get(op, 0) + 1}) for op in set(x)])
print()
```

```python
template = pd.read_csv("data/preprocessed/HDFS.log_templates.csv")
```

```python
template["FailedCountUnique"] = template["EventId"].map(lambda x: failed_operation_count_unique.get(x, 0))
template["SuccessCountUnique"] = template["EventId"].map(lambda x: success_operation_count_unique.get(x, 0))
```
